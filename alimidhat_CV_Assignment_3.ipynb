##Generating Triplet dataset
* I'm using random strategy, storing unique vehicle id (label) as a key in a dictionary, and the value is a list of which the index of that same label appearing in the train_labels .
* looping through the dict. if the value (list) of the label is less than two , skip (since i need the same label to form an anchor and positve images). else loop through the value (list) and choose anchor and positve images randomly , and for the negative vimage choose any image which doesnt have the same label.
* repeat this until i have a list of ~37k values meaning each image became an anchor at some point.


```python
# Dataset with triplet selection
"""
Anchor is a random image of a specific vehicle.
Positive is another random image of the same vehicle ID.
Negative is a random image of a different vehicle ID.
"""
import random
def generate_triplet_data(imgs, labels):
    triplets = []
    label_to_indices = {label:[] for label in set(labels)}
    # store each label position inside the dictionary
    for index, label in enumerate(labels):
      label_to_indices[label].append(index)

    for label in label_to_indices:
      pos_indices = label_to_indices[label]

      if len(pos_indices) < 2: # skip (cannot form (A,P) from less than 2)
        continue
      for i in range(len(pos_indices)):
        a, p = random.sample(pos_indices, 2) #return two pos samples (same id)
        n_label = random.choice([x for x in label_to_indices if x !=label])
        n = random.choice(label_to_indices[n_label])

        triplets.append((imgs[a], imgs[p], imgs[n]))
    return triplets
```

```python
# Example , first three images the first two (a,p) has id of 1,1 and the last one(n) has an id of 370
# Bear in mind that a,p are not the same picture (but only same label a.k.a. same car)
triplets = generate_triplet_data(train_imgs, train_labels)
triplets[:3]
```

##Triplet Loss

```python
# I used margin as 0.2 but it came back with very bad results
# Changed it to 0.5 and it enhanced the model a bit.
def triplet_loss(a,p,n, margin=0.5):
  # squared distance between anchor and positve images
  pos_dist = tf.reduce_sum(tf.square(a - p), axis=1)
  # squared distance between anchor and negative images
  neg_dist = tf.reduce_sum(tf.square(a - n), axis=1)

  loss = tf.maximum(0.0, margin + pos_dist - neg_dist)
  return tf.reduce_mean(loss)

```

##preprocess triplets

* Prepare the data of images by using tensorflow method , data.Dataset to handle the data into more suitable data structure for the model to accept.
* After that i use the preprocess image function to transform into tensors and resizing the images

```python
def preprocess_triplets(triplets, batch_size):

  anchor_imgs, pos_imgs, neg_imgs = zip(*triplets)
  # Needed to feed it to the tf.data.Dataset
  anchor_imgs = list(anchor_imgs)
  pos_imgs = list(pos_imgs)
  neg_imgs = list(neg_imgs)

  # sperating each into different datasets
  anchor_ds = tf.data.Dataset.from_tensor_slices(anchor_imgs)
  pos_ds = tf.data.Dataset.from_tensor_slices(pos_imgs)
  neg_ds = tf.data.Dataset.from_tensor_slices(neg_imgs)

  # each dataset is feed into a single major dataset
  triplet_ds = tf.data.Dataset.zip((anchor_ds, pos_ds, neg_ds))
  # Preproccessing the images before feeding it into the model
  triplet_ds = triplet_ds.map(lambda a,p,n : (preprocess_image(a), preprocess_image(p),  preprocess_image(n)))
  # return the dataset with required batch size
  return triplet_ds.batch(batch_size)

```

##Using transfer learning on ResNet50 model


```python
# cancel the classifications layers
resnet = ResNet50(weights="imagenet", include_top=False, input_shape=(224,224,3))

# freezing the first layers so it doesnt change with training
for layer in resnet.layers:
  layer.trainable = False
x = resnet.output
x = GlobalAveragePooling2D()(x)
x = Dense(512, activation="linear")(x)
x = Dense(256, activation="linear")(x)
embedding_output = Dense(128, activation="linear")(x)
# accepts 224,224 image and outputs its embedding in 128 dimensions
embedding_model = Model(inputs=resnet.input, outputs=embedding_output)
```

##Training model

* used this function because i can't used keras model.fit , this due to the fact using a custom loss fucntion (triplet loss), so i update and calculate the loss fucntion manually.
* @tf.function make custom functions like mine use tensorflow optimization so faster than normal functions.

```python
@tf.function #uses tf graph aka faster
def train_step(a, pos, neg):
  with tf.GradientTape() as tape: #manually compute the gradients
    # Forward pass
    a_embed = embedding_model(a, training=True)
    pos_embed = embedding_model(pos, training=True)
    neg_embed = embedding_model(neg, training=True)

    # calculate the loss
    loss = triplet_loss(a_embed, pos_embed, neg_embed)
  # Bp
  gradients = tape.gradient(loss, embedding_model.trainable_variables)
  optimizer.apply_gradients(zip(gradients, embedding_model.trainable_variables))

  return loss
```

* Rank all test images given a query image

```python
def rank_test_images(query_embedding, test_embeddings):
    # Calculate the Euclidean distance between query and all test embeddings
    distances = tf.norm(test_embeddings - query_embedding, axis=1)
    # Get indices of sorted distances (smallest distance is most similar)
    sorted_indices = tf.argsort(distances)
    return sorted_indices

ranked_indices_all_queries = []
for query_embedding in query_embeddings:
    ranked_indices = rank_test_images(query_embedding, test_embeddings)
    ranked_indices_all_queries.append(ranked_indices)
```

## mAP
* loop through each query label and then over its ranked items, at each point compare the id of both the rank (test) and query images if its the same label add it to the relevancy list.
* For each relevant rank, it calculates the precision, which is the ratio of relevant items to the total number of items examined up to that rank.
* the avg precision is the mean of the preivous values (to all relevent ranks).
* Then calculates the mean for all queries which is the mAP.
